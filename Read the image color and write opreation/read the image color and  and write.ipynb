```python
import cv2
import os
```

# Read Image

```python
img_path = os.path.join('F:\computer vision project\OpenCV-Basics-main\APJ.jpg')
```

```python
img_path
```

```python
img=cv2.imread(img_path)
```

```python
img
```

# Going to render the image

```python
from matplotlib import pyplot as plt
```

```python
#Show image
plt.imshow(img)
```

```python
#i am going to recolor the image
recolor = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
recolor
```

```python
plt.imshow(recolor)
```

```python
# now i am going to color in gray
grayimg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grayimg)
```

# Write out image

```python
cv2.imwrite('grayimga.jpg',grayimg)
```
